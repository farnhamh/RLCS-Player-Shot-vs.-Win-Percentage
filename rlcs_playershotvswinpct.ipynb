```python
import pandas as pd

#player data for every game in the 2021-22
dfPlayers = pd.read_csv('games_by_players.csv', dtype='string')

#used to replace team_id with the team name
dfTeams = pd.read_csv('games_by_teams.csv', dtype='string')

#used to clean dfPlayers to only events listed as a "Major" or "World Championship"
dfMain = pd.read_csv('main.csv', dtype='string')

#cleans dfPlayer to only events listed as a "Major" or "World Championship"
dfMain = dfMain[(dfMain['event'] == 'Major') | (dfMain['event'] == 'World Championship')]
gameData = []
for i in range(len(dfMain)):
    mainId = dfMain.iloc[i]['game_id']
    if mainId in dfPlayers['game_id'].values:
        gameData.append(dfPlayers.loc[dfPlayers['game_id'] == mainId])
dfGames = pd.concat(gameData)
dfGames
```

```python
#replaces team_id in df with the team name
dfTeams = dfTeams.drop_duplicates('team_id')

teamDict = {}
teamNames = []
win = []
loss = []
for i in range(len(dfTeams)):
    teamDict[dfTeams.iloc[i]['team_id']] = dfTeams.iloc[i]['team_name']

for i in range(len(dfGames)):
    teamNames.append(teamDict[dfGames.iloc[i]['team_id']])
    if dfGames.iloc[i]['winner'] == 'False':
        win.append(0)
        loss.append(1)
    else:
        win.append(1)
        loss.append(0)
dfGames['team_name'] = teamNames
dfGames['win'] = win
dfGames['loss'] = loss
df = dfGames.reset_index()
df= df[['game_id', 'team_name', 'player_tag', 'core_shots', 'core_goals', 'win', 'loss']]
df['core_shots'] = pd.to_numeric(df['core_shots'])
df
```

```python
def playerShotWinPct (player):
    playerDf = df[df['player_tag'] == player]
    winCount = playerDf.groupby('core_shots').agg('sum').reset_index()
    winPct = []
    for i in range(len(winCount)):
        winPct.append(winCount.iloc[i]['win']/(winCount.iloc[i]['win'] + winCount.iloc[i]['loss']))
    winCount['winPct'] = winPct
    return winCount
playerShotWinPct('Darth.')
```

```python
dfPlayers = df.groupby('player_tag').agg('count').reset_index()
playerRegressionDf = dfPlayers[['player_tag']]
playerRegressionDf
winSlope = []

from sklearn.linear_model import LinearRegression

for player in playerRegressionDf['player_tag']:
    dfSlope = playerShotWinPct(player)
    model = LinearRegression()
    model = model.fit(dfSlope[['core_shots']], dfSlope['winPct'])
    winSlope.append(float(model.coef_))
playerRegressionDf['winSlope'] = winSlope
playerRegressionDf.sort_values(by=['winSlope'], ascending=False).head(10)
```
